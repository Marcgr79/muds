Los cuatro data sets parecen ser por los estadísticos calculados muy parecidos, los datos no son exactamente los mismos aunque los resultados si lo son.

```python
#Visualización

import matplotlib.pyplot as plt

def scatter(dataframe, x_column, y_column, title="", color="blue", alpha=0.5):
    plt.scatter(dataframe[x_column], dataframe[y_column], color=color, alpha=alpha)
    plt.title(title)
    plt.show()

scatter(dataset1, x_column='x', y_column='y', title='Scatter Plot 1')
scatter(dataset2, x_column='x', y_column='y', title='Scatter Plot 2')
scatter(dataset3, x_column='x', y_column='y', title='Scatter Plot 3')
scatter(dataset3, x_column='x', y_column='y', title='Scatter Plot 4')
```

Una vez visualizados los datos, se confirma que no son los mismos, si tiene que coincide que tienen los mismos resultados estadísticos pero no son los mismos datos. Solo hay que fijarse en la visualización de todos estos datos para entender que los datos son diametralmente distintos.

La paradoja de Anscombes nos esta desmotrando aquí que nuestra cabeza ha querido confirmar una idea (confirmation bias) porque todos los indicios apuntaban hacia esa, sin antes intentar entender mejor la información que nos estaban dando. Un buen data scienst, y no es la primera vez que nos lo dicen en el master, debe siempre visualizar los datos y entenderlos a fondo, para luego empezar a tratarlos. Al tener unos resultados tan evidentes (e ir tan justos de tiempo) hemos querido confirmar algo que realmente no decidíamos.

En mi defensa dire que si he provado de hacer una confirmación, he mirado las tablas, y al ser cortas, he visto que los datos no eran los mismos. Por eso he añadido la parte "... los datos no son exactamente los mismos aunque los resultados si lo son."
