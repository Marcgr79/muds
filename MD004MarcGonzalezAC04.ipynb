## 2. Creación del Modelo

Antes de empezar, vamos a tener que dividir los datos que usaremos para generar este modelo, generaremos una semilla para que siempre se haga la misma partición y para que las diferencias entre los casos no se deban a que se hacen selecciones de datos distintos para cada uno de los grupos. Y una vez divididos los datos, podremos crear el modelo.

```r
install.packages('caret')
library(caret)
```

```r
set.seed(7)

Train = createDataPartition(y = data$target_deathrate,
                            p = 0.75,
                            list = FALSE,
                            times = 1)
data_train = data[Train, ]
data_test = data[-Train, ]
```

```r
nrow(data_train)
nrow(data_test)
```

Esto es una comprovación rápida de que la partición se ha hecho correctamento. Lo que estamos calculando es la cantidad de datos que hemos almacenado en cada uno de los grupos.

**Modelo Incidence Rate:**

```r
model_incidencerate = lm(target_deathrate ~ incidencerate , data = data_train)
summary(model_incidencerate)
```

Para empezar, aquí creamos el modelo, definimos nuestra variable objetivo como target death rate y nuestra variable explicativa como incident rate. Luego fijamos que los datos que usaremos son los que surgen de la partición data train y observamos los resultados.

El intercepto es de 83,35 significando que el resultado de la regresión cuando la variable explicativa sea 0 tomará este valor. Por otro lado, la B1 es de 0,21.

Luego también observamos el nivel de impacto o explicación que nos da este modelo sobre los cambios en la variable objetivo, esto se puede observar en la R-squared que esta cerca de 0,17, es decir, que el modelo nos explica un 17% de la variable objetivo.

```r
overview = summary(model_incidencerate)
```

```r
#Normalidad
#Gráfico
qqnorm(overview$residuals, col = 'darkgreen')
qqline(overview$residuals)
```

Aquí comprobamos el primero supuesto. Observamos si los residuos del modelo siguen una distribución normal con el gráfico QQ-plot. Podemos ver que tanto al inicio, como al final, hay bastantes residuos que se desvían de una forma notable de la línea que nos marca la normalidad. A continuación veremos dos tests, tanto el Anderson-Darling como el Shapiro-Wilk, que nos confirman esta teoría, ambos rechazan la hipótesis nula de que estos residuos siguen la distribución normal.

A pesar de todo esto, uno podría llegar a argumentar que con una distribución de aproximadamente 2000 observaciones (residuos) se podría sospechar que tenderán a una distribución normal. Sería poco riguroso, pero en el caso de que en algún modelo se cumplan todos los supuestos y falle este, sería una posibilidad a explorar.

```r
install.packages('nortest')
library(nortest)
```

```r
#Normalidad
#Anderson-Darling Test
normalidad_residuals = ad.test(overview$residuals)
print(normalidad_residuals)
```

```r
#Normalidad
#Shapiro-wilk Test
shapiro.test(overview$residuals)
```

```r
install.packages('lmtest')
library(lmtest)
```

```r
#Homocedasticidad
#Test Breusch-Pagan
bptest(model_incidencerate)
```

```r
#Grafico de dispersión
plot(model_incidencerate$fitted.values, model_incidencerate$residuals, pch =19, col ="green", cex=2 )
```

Aquí lo que he hecho ha sido un test para comprobar la homoscedasticidad de los errores. Se define esta característica como una propiedad deseable en los modelos de regresión, que analiza la dispersión y la constancia entre los errores analizados. Esta propiedad de un modelo aumentará su fiabilidad.

Con el test de Breusch-Pagan definimos la hipótesis nula como que el modelo tiene unos residuos homocedastícos y, pese a darnos un p-value no demasiado alejado de la zona de aceptación, tenemos que rechazar esta hipótesis nula.

Además, he querido ver esto con un gráfico de dispersión con los valores previstos y los residuos, deberíamos ver una distribución aleatoria. En cambio, nosotros vemos una nube de puntos muy centrada en una parte de nuestro gráfico, con un outlier en la parte inferior derecha y sin ninguna forma similar a lo largo del rango de observaciones.

```r
#Autocorrelación
#Durbin-Watson Test
dwtest(model_incidencerate, alternative = c('two.sided'))
```

```r
plot(model_incidencerate$residuals, pch =19,col ="green", cex=2)
```

Por último, vamos a comprobar la autocorrelación de estos errores. También llamada independencia, la autocorrelación intenta analizar la independencia de los errores entre si, para poder afirmar que no hay correlación entre ellos y tener así un modelo de regresión lo más fiable posible.

En este caso hacemos el Durbin-Watson test donde la hipótesis nula afirma que, efectivamente, la correlación entre los errores es de 0. Como podemos observar, en este caso esto no pasa, ya que rechazamos esta hipótesis.

Además, al graficar los residuos, también se puede observar una correlación entre ellos. Si nos fijamos en el gráfico de dispersión vemos que muchos de ellos están distribuidos alrededor del valor 0 de este gráfico siguiendo un patrón similar. Seguimos teniendo un outlier claro.

```r
#Media
mean(overview$residuals)
```

*Modelo Lineal Basado en la Variable Incidence Rate:*

Podemos observar que es un modelo que nos da un 17% de la informacióbn sobre la variable objetivo, ya que tenemos que la R-squared es igual a 0,17.

En cambio, y esto lo podemos ver en las siguientes líneas de código que he desarrollado, es un modelo que no cumple con la mayoría de los supuestos que deben cumplir los modelos de regresión lineal. # 3. Validación del Modelo

Una vez llegados a este punto, tenemos que buscar el modelo más fiable, o en nuestro caso, el 'menos malo' y probar de hacer las predicciones. Para ello vamos a usar los conjuntos de datos data_test correspondientes que hemos creado antes de generar cada modelo.

Una vez hechas las predicciones, vamos a visualizar los datos de nuestro modelo alrededor de la predicción que haría nuestro modelo para comparar la predicción con la realidad.

Después de comparar todos los resultados obtenidos en los anteriores pasos, el modelo que parece que será más fiable es el modelo con la doble transformación de variables y con la variable explicativa Incidence Rate. Se observa que este modelo está muy muy cerca de aceptar el supuesto de normalidad, es el único modelo en el que se acepta la homoscedasticidad y el modelo con una mediana de los errores más pequeña. Así que será el que probaremos para predecir.

Veámoslo:

```r
prediccion_incidence_sqrt2 = predict(model_incidence_sqrt, data_test2)
```

```r
#Graficamos
plot(data_test2$target_deathrate, prediccion_incidence_sqrt2, xlab = 'Target_Test', ylab = 'Predicción',  pch = 15, col = 'lightgreen')

lm.out = lm(prediccion_incidence_sqrt2 ~ target_deathrate, data = data_test2)
newx = seq(min(data_test2$target_deathrate),max(data_test2$target_deathrate),by = 0.05)
conf_interval = predict(lm.out, newdata=data.frame(target_deathrate=newx), interval="confidence", level = 0.95)
abline(lm.out, col = "black")

lines(newx, conf_interval[,2], col="darkgrey", lty=2)
lines(newx, conf_interval[,3], col="darkgrey", lty=2)
```

Una vez hecha la predicción, al visualizar los datos obtenemos un poco lo esperable. Un modelo que sí que parece seguir la tendencia de los datos reales con claridad, pero que tiene muchos datos que se desvían en exceso de la regresión. Pese a que podemos observar que el grueso de los datos sigue el patrón de la regresión de los datos reales, muchos de nuestros resultados quedan fuera de los intervalos de confianza.

```r
data.frame( R2 = R2(prediccion_incidence_sqrt2, data_test2$target_deathrate),
            RMSE = RMSE(prediccion_incidence_sqrt2, data_test2$target_deathrate),
            MAE = MAE(prediccion_incidence_sqrt2, data_test2$target_deathrate))
```

En relación con las medidas estadísticas, obtenemos una R2 de 0,29, indicando que un 29% de la variabilidad de nuestra variable objetivo (target_deathrate) se explica en el modelo. No es un mal número teniendo en cuenta las muchísimas variables que tenemos, pero sí que está bastante alejado de 1 que el que nos indicaría un buen ajuste en el modelo y que tenemos una variable explicativa muy bien definida.

Por otro lado, tenemos una RMSE y una MAE que nos explican que nuestras predicciones difieren en 0,95 de la variable respuesta de media. I un error absoluto medio (MAE) que nos dice que la diferencia en promedio entre las predicciones y la variable objetivo es de 0,73.
